```python
img= cv.imread("/content/drive/My Drive/Colab Notebooks/IMAGES/logo/logo.png")
img =cv.cvtColor(img,cv.COLOR_BGR2RGB)
resize_img=cv.resize(img,(64,64))

cv.imwrite("/content/drive/My Drive/Colab Notebooks/resize/logo.png", resize_img)
```

```python
img2= "/content/drive/My Drive/Colab Notebooks/resize"
images=os.listdir(img2)
len(images)
```

```python
img_data=[]
for img in images:
  img_arr=cv.imread(os.path.join(img2,img))
  img_data.append(img_arr)
plt.figure(figsize=(10,10))
for i in range(len(img_data)):
  plt.subplot(6,4,i+1)
  plt.imshow(img_data[i])  
```
